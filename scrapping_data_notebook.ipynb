# Scrapping data from pocoso website 

* I am going to write a code which downloads all the weekly data pdf's 
* There are 434 pdf's present in the website , I am going to download all of them at once
* I am going to get the required table from the pdf's we downloaded and saved 
* There are some situations where , the table we want is in an image format 

```python
import requests
import bs4
from bs4 import BeautifulSoup as bs
import urllib.request
from urllib.request import urlopen as uReq
import re 

posoco_url = "https://posoco.in/reports/weekly-reports/"# url which contains the yearwise links (those links direct to weekly data pdf's) 
uClient = uReq(posoco_url)                              # requesting the webpage from the internet
PosocoPage = uClient.read()                             # reading the webpage
uClient.close()                                         # closing the connection to the web server
Posoco_html = bs(PosocoPage, "html.parser")             # parsing the webpage as HTML
bigboxes = Posoco_html.findAll("div", {"class": "entry-content"}) # searching for appropriate tag to retrieve the links
first_links=[]
for beta in bigboxes[0].findAll('a'):
    first_links.append(beta['href'])                    #extracted all the link and appended to first_links
# print(first_links)
# print('=========================================')
all_links=[]
for link in range(len(first_links)):
    posoco_url1 = first_links[link]
    uClient = uReq(posoco_url1)                         # requesting the webpage from the internet
    PosocoPage1 = uClient.read()                        # reading the webpage
    uClient.close()                                     # closing the connection to the web server
    Posoco_html1 = bs(PosocoPage1, "html.parser")       # parsing the webpage as HTML
    bigboxes1 = Posoco_html1.findAll("div", {"class": "container-fluid"}) # href link is in the class container-fluid
    for i in bigboxes1[0].findAll('tr'):
        for x in i.findAll('a'):                        # finding a tag to get href 
            all_links.append(x['href'])
print('Total no of pdfs available in the posoco website is {size}:'.format(size=len(all_links)))
print('==========================================')
count_file=0
filenames_list=[] # we are getting all the filenames into list , so that we can use them back for extracting the tables
for links in all_links:
    count_file+=1
    filename = str(re.findall('d/(.+)/',links)[0])+'.pdf'
    filenames_list.append(filename)
    urllib.request.urlretrieve(links, filename)
    print('saved pdf{count} from this link {li}'.format(count=count_file,li=links))
    
print('success :)')
```

Now we are going to use tabula and pyPDF2 libraries to extract the tables from all the pdfs

```python
import tabula
import PyPDF2 

all_table=[]
count=0
for file_name in filenames_list:
    if file_name not in list_nodata:                  
        pdfFileObj = open(file_name, 'rb')             # creating a pdf file object
        count+=1
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj,)  # creating a pdf reader object
        pages = pdfReader.numPages                     # number of pages in pdf file
        for i in range(1,pages):
            pageObj = pdfReader.getPage(i)             # creating a page object
            text = pageObj.extractText()               # extracting text from page
            if 'Energy Consumption in States (MUs)' in text:
                # We only want tables who's heading is 'Energy Consumption in States (MUs)'
                # We will find that heading and get that page number where the table is located 
                required_page = i
                # we got the page number
                break
            else:
                continue
        table = tabula.read_pdf(file_name,pages=required_page+1)
        try:                                            # try to extract table from the pdf
            table = table[0].drop(table[0].tail(1).index)
            all_table.append(table)
        except:                                         # else print the file name
            print(table)
            print(count)
            print(file_name)
            all_table.append(table)
#         print(table)        
#         print('=====================')
```
